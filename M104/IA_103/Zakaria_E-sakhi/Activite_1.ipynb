```python
#Reshape image matrix from 24×24 into 784
x_train = x_train.reshape(60000,784)
x_test= x_test.reshape(10000,784)
```

Reshaping  from a matrix into an a vector :

- The primary cause is basic Neural Networks & Classical ML (Dense/Fully Connected Layers)
Standard algorithms (like Linear Regression, Random Forests, or basic Multi-Layer Perceptrons) cannot process a 2D grid of pixels. They expect a flat table where each row is a sample and each column is a single feature

-as a result the whole picture become a vector

```python
#Ex of a picture as a vector

x_train[4]
```

```python
#scaling
from sklearn.preprocessing import MinMaxScaler
Scaler = MinMaxScaler().fit(x_train)
x_train_scaled = Scaler.transform(x_train)
x_test_scaled = Scaler.transform(x_test)
```

```python
#initialise the model with 3 hidden layers and an output layer
input_shape = x_train.shape[1]

model = keras.Sequential([
    keras.layers.Dense(300, activation='relu', input_shape=(input_shape,)),
    keras.layers.Dense(300, activation='relu'),
    keras.layers.Dense(100, activation='relu'),
    keras.layers.Dense(10, activation='softmax'),
])

model.compile(optimizer='sgd',
              metrics=["accuracy"],
              loss="sparse_categorical_crossentropy")

history = model.fit(x_train_scaled,y_train,
             #batch_size=256,
             epochs=5,
             validation_split =0.1)
```

```python
model.summary()
```

```python
import pandas as pd

history_df =pd.DataFrame(history.history)

history_df[["loss","val_loss"]].plot(title="The train loss vs validation loss")
```

```python
history_df[["accuracy","val_accuracy"]].plot()
```

#No overfitting detected in result so no need for regularization

```python
#test model performance
import numpy as np
from sklearn.metrics import classification_report
prediction_probabilities = model.predict(x_test_scaled)
prediction_labels = np.argmax(prediction_probabilities, axis=1)
print(classification_report(y_test, prediction_labels))
```
